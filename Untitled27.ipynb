```python
import pymongo
import os
from llama_index import ServiceContext, GPTVectorStoreIndex, LLMPredictor, PromptHelper, SimpleDirectoryReader, load_index_from_storage
from langchain import OpenAI
from llama_index.node_parser import SimpleNodeParser
from llama_index import StorageContext, load_index_from_storage

# Set the OpenAI API key
os.environ["OPENAI_API_KEY"] = "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"  # Replace with your OpenAI API key

# Establish a connection to the guidance database
mongo_client = pymongo.MongoClient("mongodb+srv://nitik:nitik123@cluster0.2e6kk1k.mongodb.net/?retryWrites=true&w=majority")
database = mongo_client['database']
collection = database['collection']

# Initialize the node parser
parser = SimpleNodeParser()

def construct_index(directory_path):
    max_input_size = 4096
    num_outputs = 500
    chunk_overlap_ratio = 0.1  # Adjust the value between 0 and 1
    chunk_size_limit = 1024

    print("*" * 5, "Documents parsing initiated", "*" * 5)
    file_metadata = lambda x: {"filename": x}
    reader = SimpleDirectoryReader(directory_path, file_metadata=file_metadata)
    documents = reader.load_data()

    prompt_helper = PromptHelper(max_input_size, num_outputs, chunk_overlap_ratio, chunk_size_limit=chunk_size_limit)
    llm_predictor = LLMPredictor(llm=OpenAI(temperature=0, model_name="gpt-3.5-turbo", max_tokens=num_outputs))

    service_context = ServiceContext.from_defaults(llm_predictor=llm_predictor, prompt_helper=prompt_helper)

    index = GPTVectorStoreIndex.from_documents(
        documents=documents, service_context=service_context
    )

    index.storage_context.persist("./entire_docs")
    return index

construct_index("./docs")
storage_context = StorageContext.from_defaults(persist_dir="./entire_docs")
index = load_index_from_storage(storage_context)
query_engine = index.as_query_engine()

while True:
    text_input = input("YOU : ")

    # Retrieve guidance result from the database
    guidance_result = collection.find_one({'query_text': text_input})
    if guidance_result:
        guidance_text = guidance_result['guidance_text']
        text_input += " " + guidance_text

    response = query_engine.query(text_input)
    print("Bot : ", response)
    print('\n')
```
